**Applying Logistic Regression Model**

```python
# X is Input Variable
x = df.iloc[:,0:13].values
x
```

```python
# Y is Output Variable
y = df.iloc[:,13].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20, random_state = 0)
```

```python
#SCALING OR NORMALIZATION
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
#predictor variable/predict the output
y_pred = model.predict(x_test)
y_pred   #predicted output values
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)* 100
```

```python
from sklearn.pipeline import Pipeline
dep_model = Pipeline([('minmax',MinMaxScaler()), ('LR',LogisticRegression())])
dep_model.fit(x_train,y_train)
```

```python
import joblib
joblib.dump(dep_model,'HeartDiseasePrediction')
#Dumping the deployed/Pipelined model in HeartDiseasePrediction.csv file
```
